# Using the MTCARS data set: 

- 1) Apply backward elimination to find top 3 input variables which impact the mileage of the car

- 2) Apply factor analysis to see which columns can be grouped together

```python
import pandas as pd
```

```python
mtcars = pd.read_csv('F:\Skillenable\Data frames\mtcars.csv')
mtcars.head()
```

```python
cars= mtcars.rename(columns={'Unnamed: 0':'Cars'})
```

```python
#  backward elimination
# Step 1: Find your X and Y

Y = cars[['mpg']]
X = cars.drop(columns='mpg')
```

```python
# Step 2: # spliting data into training and test

from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size = 0.8, random_state = 1234)

len(X_train), len(X_test), len(Y_train), len(Y_test)
```

```python
# Step 3: Create the black linear Regration object

from sklearn.linear_model import LinearRegression
 
lm = LinearRegression()
```

```python
# step 4: Find the set of effective input variable (Feature selection)

from mlxtend.feature_selection import SequentialFeatureSelector as sfs
column_name = sfs(lm, k_features= 3, forward=False, verbose= 1)
```

```python
# step 6: find  name of top 3 input column
final_sfs = column_name.fit(X,Y)
list(final_sfs.k_feature_names_)
```

## 'Cars', 'cyl', 'disp' are top 3 input variable.

# Factor analyser

```python
# import required functions
from factor_analyzer import FactorAnalyzer
from factor_analyzer.factor_analyzer import calculate_kmo
```

```python
# drop na columns
cars_fa = cars.drop(columns='Cars')
cars_fa
```

```python
# KMO test
kmo_all, kmo_model = calculate_kmo(cars_fa)
print(kmo_model)
```

```python
# Let's create 10 factors  to find the right number of factors inwhich this data can be divided

# step 1: Creating a factor object
fa = FactorAnalyzer(n_factors = 10, rotation = None)
```

```python
#step 2: Fit the object on data frame
fa.fit(cars_fa)
```

```python
# step 3: Find eigen value
eigen_values, eigen_values = fa.get_eigenvalues()
list(eigen_values)
```

```python
# creating model for n_factor = 2
fa_final = FactorAnalyzer(n_factors=2, rotation = 'varimax')
```

```python
# fitting the object
fa_final.fit(cars_fa)
```

```python
# printing loading factor
fa_final.loadings_
```

```python
#have a dataframe
loading_df = pd.DataFrame(fa_final.loadings_, columns=['Factor1', 'Factor2'], index = cars_fa.columns)
loading_df
```
